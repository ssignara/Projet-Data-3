```python
import pandas as pd
import numpy as np
import matplotlib as mp
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
pop = pd.read_csv("population.csv") 
aa = pd.read_csv("aide_alimentaire.csv")
da = pd.read_csv("dispo_alimentaire.csv")
sn = pd.read_csv("sous_nutrition.csv")
cer = pd.read_csv('céréales.csv')
```

```python
# Calcul de la population mondiale en 2017

pop_2017_df = pop[pop["Année"] == 2017]
pop_2017 = pop_2017_df['Valeur'].sum()

print('La population mondiale en 2017 est de', '{:,}'.format(int(pop_2017*1e3)).replace(',', ' '), 'habitants.')
```

```python
# Calcul du nombre de personnes en état de sous-nutrition en 2017

sn_2017_df = sn[sn["Année"] == '2016-2018']

sn_2017_df.loc[sn_2017_df['Valeur'] == '<0.1'] = 0
sn_2017_df['Valeur'] = sn_2017_df['Valeur'].apply(pd.to_numeric)

sn_2017 = sn_2017_df['Valeur'].sum()
print(int(sn_2017),'Millions de personnes étaient en état de sous-nutrition en 2017.')
```

```python
# Marc - Question n°1 : Proportion de personnes en état de sous-nutrition en 2017

print('La proportion de la population mondiale en état de sous-nutrition pour l\'année 2017 est de',round((sn_2017*1e3)/(pop_2017)*100, 3),'%.')
```

```python
# Jointure des tableaux 'Disponibilité alimentaire' et 'Population' en 2017

pop_da = pd.merge(pop_2017_df, da, on='Zone')
pop_da
```

```python
# Marc - Question n°2 : Nombre théorique de personnes qui pourraient être nourries en 2017 (à partir de la disponibilité alimentaire mondiale)

# D'après l'OMS, il est indiqué que les besoins caloriques journaliers sont de 2500kcal/jour soit 912500kcal/an.

pop_da_kcal = round(((pop_da['Disponibilité alimentaire (Kcal/personne/jour)']*365*pop_da['Valeur']).sum()/912500)*1e3, 2)

pop_da_kcal_percent = round(((pop_da_kcal/(pop_2017*1e3))*1e2),2)

print('La disponibilité alimentaire mondiale permettrait de nourrir ', '{:,}'.format(int(pop_da_kcal)).replace(',', ' '), 'personnes, soit ', pop_da_kcal_percent, '% de la population mondiale.')
```

```python
# Création du filtre afin de n'avoir que les disponibilités végétales

pop_da_veg = pop_da[pop_da['Origine'] == 'vegetale']
pop_da_veg
```

```python
# Marc - Question n°3 : Nombre théorique de personnes qui pourraient être nourries en 2017 (à partir de la disponibilité alimentaire mondiale en végétaux)

pop_da_veg_kcal = round(((pop_da_veg['Disponibilité alimentaire (Kcal/personne/jour)']*365*pop_da_veg['Valeur']).sum()/912500)*1e3, 2)

pop_da_veg_kcal_percent = round(((pop_da_veg_kcal/(pop_2017*1e3))*1e2),2)

print('La disponibilité alimentaire mondiale en végétaux permettrait de nourrir ', '{:,}'.format(int(pop_da_veg_kcal)).replace(',', ' '), 'personnes, soit ', pop_da_veg_kcal_percent, '% de la population mondiale.')
```

```python
# Marc - Question n°4 : Utilisation de la disponibilité intérieure, en particulier la part qui est attribuée à l’alimentation animale, celle qui est perdue et celle qui est concrètement utilisée pour l'alimentation humaine en 2017

# Utilisation de la disponibilité intérieure
# Disponibilité intérieur = Production + importations - exportations + variations des stocks = Nourriture + Aliments pour animaux + Semences + Traitement + Autres utilisations + Pertes

# Part qui est attribuée à l’alimentation animale
p1 = round(da['Aliments pour animaux'].sum()/da['Disponibilité intérieure'].sum()*100, 2)
print(p1 ,'% de la disponibilité intérieure sont attribués à l\'alimentation animale.')

# Part qui est perdue
p2 = round(da['Pertes'].sum()/da['Disponibilité intérieure'].sum()*100, 2)
print(p2 ,'% de la disponibilité intérieure correspondent aux pertes.')

# Part qui est concrètement utilisée pour l'alimentation humaine
p3 = round(da['Nourriture'].sum()/da['Disponibilité intérieure'].sum()*100, 2)
print(p3 ,'% de la disponibilité intérieure sont attribués à l\'alimentation humaine.')
```

```python
# Julien - Question n°1 : Répartition entre l’alimentation humaine et l’alimentation pour animaux

# Création du filtre sur les céréales
liste_cer = cer['Produit'].unique()
da_cer_df = da.loc[da['Produit'].isin(liste_cer),:]

# Calcul de la disponibilité intérieure en céréales
da_cer = da_cer_df['Disponibilité intérieure'].sum()

# Calcul de la part destinée à l'alimentation animale
da_cer_a = da_cer_df['Aliments pour animaux'].sum()
print(round(da_cer_a/da_cer * 100, 2), '% de la disponibilité intérieure mondiale en céréales est destinée à l\'alimentation animale.')

# Calcul de la part destinée à l'alimentation humaine
da_cer_h = da_cer_df['Nourriture'].sum()
print(round(da_cer_h/da_cer * 100, 2), '% de la production mondiale de céréales est destinée à l\'alimentation humaine.')
```

```python
# Création du filtre sur le manioc et projection sur la Thaïlande

da_thai_manioc = da[(da['Zone'] == 'Thaïlande') & (da['Produit'] == 'Manioc')]
da_thai_manioc = da_thai_manioc[['Zone', 'Produit', 'Exportations - Quantité', 'Production']]
da_thai_manioc
```

```python
# Julien - Question n°2 : Utilisation du manioc par la Thaïlande aux égards de la proportion de personnes en sous-nutrition et calcul de l’exportation par rapport à la production

# Définition du nombre de personnes estimée comme étant en sous-nutrition en Thaïlande pour l'année 2017
sn_thai_2017_df = sn_2017_df[['Zone', 'Année', 'Valeur']][(sn_2017_df['Zone'] == 'Thaïlande')]
sn_thai_2017 = sn_thai_2017_df['Valeur'].sum()

# Définition du nombre de personnes vivant en Thaïlande en 2017
pop_thai_2017_df = pop_2017_df[['Zone', 'Année', 'Valeur']][(pop_2017_df['Zone'] == 'Thaïlande')]
pop_thai_2017 = pop_thai_2017_df['Valeur'].sum()

# Calcul de l’exportation par rapport à la production
print('La Thaïlande a exporté ', round(25214/30228*100, 2), '% de sa production de manioc en 2017 alors que, dans le même temps, ', round(sn_thai_2017/pop_thai_2017*1e5, 2), '% des thaïlandais étaient considérés comme étant en sous-nutrition.')
```

```python
# Jointure des tableaux 'Sous-nutrition' et 'Population' en 2017

pop_sn = pd.merge(pop_2017_df, sn_2017_df, on='Zone')
pop_sn = pop_sn.replace(np.nan, 0)
pop_sn
```

```python
# Mélanie - Question n°1 : Liste des pays pour lesquels la proportion de personnes sous-alimentées est la plus forte en 2017

# Création des colonnes 'Population' et 'Population en sous-nutrition' en Millions
pop_sn['Population (en Millions)'] = pop_sn['Valeur_x']*1e-3
pop_sn['Population en sous-nutrition (en Millions)'] = pop_sn['Valeur_y']

del pop_sn['Valeur_x']
del pop_sn['Valeur_y']

# Création de la colonne 'Pourcentage de la population en sous-nutrition' et classement dans l'ordre décroissant
pop_sn['Pourcentage de la population en sous-nutrition'] = round(pop_sn['Population en sous-nutrition (en Millions)']/pop_sn['Population (en Millions)']*100, 2)
pop_sn = pop_sn.sort_values(by='Pourcentage de la population en sous-nutrition',ascending=False)

# Création du filtre permettant d'obtenir la liste de tous les pays pour lesquels le taux de sous-nutrition est supérieur ou égal à 20%
pop_sn = pop_sn[pop_sn['Pourcentage de la population en sous-nutrition'] >= 20.00]

pop_sn
```

```python
# Mélanie - Question n°2 : Liste des pays qui ont le plus bénéficié d’aide depuis 2013

# Renommage des colonnes 'Valeur' et 'Pays bénéficiaire'
aa.rename(columns = {'Valeur':'Quantité donnée (en tonnes)', 'Pays bénéficiaire':'Zone' }, inplace=True)

# Groupage sur les zones et classement dans l'ordre décroissant afin d'obtenir les pays ayant perçu le plus d'aide entre 2013 et 2017
aa_2013_2017 = aa[['Zone','Quantité donnée (en tonnes)']].groupby('Zone').sum()
aa_2013_2017.sort_values(by='Quantité donnée (en tonnes)', ascending=False).head(10)
```

```python
# Mélanie - Question n°3 : Liste des pays ayant le plus de disponibilité/habitant (en kcal)

# Création du dataframe permettant de déterminer la disponibilité par habitant totale, groupée par zone
da_df = da[['Zone','Disponibilité alimentaire (Kcal/personne/jour)']].groupby('Zone').sum()
da_df.sort_values(by='Disponibilité alimentaire (Kcal/personne/jour)', ascending=False).head(10)
```

```python
# Mélanie - Question n°4 : Liste des pays ayant le moins de disponibilité/habitant (en kcal)

da_df.sort_values(by='Disponibilité alimentaire (Kcal/personne/jour)', ascending=False).tail(10)
```

```python
# À titre d'information, comparons les disponibilités alimentaires du pays en ayant le plus et celui en ayant le moins

da_df = da[['Zone','Disponibilité alimentaire (Kcal/personne/jour)']].groupby('Zone').sum()

da_pop = pd.merge(pop_2017_df, da_df, on='Zone').sort_values(by='Disponibilité alimentaire (Kcal/personne/jour)', ascending=False)
da_pop
```
